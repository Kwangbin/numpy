### linspace 
숫자에 선형성을 부여 한다

```python
# 1-10 까지 5개의 숫자를 선형성(직선 증가 감소의 1차원 선형)

np.linspace(0,10, num=5)
```

```python
np.linspace(0,10, num=10)
```

```python
np.linspace(5,-4, num=10)
```

```python
np.ones([2,3],dtype=np.int64)
```

###배열 생성 속도

```python
np.zeros(5)
```

```python
import time
```

```python
%timeit np.zeros(5)
```

```python
%timeit np.ones(5)
```

```python
%timeit np.empty(5)
```

```python
from numpy.core.function_base import linspace
%timeit np.linspace(1,10,5)
```

```python
%timeit np.arange(5)
```
