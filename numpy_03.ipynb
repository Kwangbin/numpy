###함수 적용시 inplace=True  매개변수 없으면 원데이터 수정 불가 있으면 수정 가능[링크 텍스트]
### inplace=True 있으면 a.function
### inplace=True 없으면 a=a.function

```python
a=a.reshape([3,4])
a
```

```python
a.shape
```

```python
#정렬의 기본 축은 열기준 axis=-1 
# -1이란 마지막 값이 있는 차원 이다
# 행, 열 이니 열이 마지막 이니 열기준

np.sort(a)
```

```python
# 0 행기준
np.sort(a, axis=0)
```

```python
# 열기준 역순 정렬
# ::-1 1차원 역순 정렬
np.sort(a)[::-1]
```

```python
# 2차원 역순 정렬은 차원을 적용 해야 한다
# :. ::-1

np.sort(a)[:, ::-1]
```

```python
# 열기준 역순 정렬
# :,::-1 1차원 역순 정렬
# 행기준 역순 정렬
# ::-1,: 1차원 역순 정렬

np.sort(a,axis=0)[::-1,:]
```

#argsort

```python
np.random.seed(42)
a =np.random.randint(0, 10, 12).reshape([3,4])
a
```

```python
np.argsort(a)
```

[6,3,7,4] 의 

argsort는 [1,3,0,2] 인덱스 값을 확인해 보자

0번 6, 1번 3, 2번 7,  3번 4 

값의 인덱스로 정력 하면 1,3,0,2 다

즉, 정렬된 값의 index(주소) 값으로 표시한다.

```python
np.sort(a)
```

##argsort는  정렬후 값이 아니 해당 값의 정렬 순위 index 값을 반환 한다

```python
np.argsort(a)[:,::-1]
```

```python
np.argsort(a)[::-1,:]
```

```python
#  2차원 배열 합치기
x = np.array([[1,2] , [3,4]])
y= np.array([5,6])
```

```python
x.shape , y.shape
```

```python
np.concatenate((x,y),axis=0)
```

에러 발생 이유눈 x는 2차원 y는 1차원 이라 차원이 맞치 않음

```python
#  2차원 배열 합치기
# y 2차원 변경
x = np.array([[1,2] , [3,4]])
y= np.array([[5,6]])
```

```python
x.shape , y.shape
```

```python
np.concatenate((x,y),axis=0)
```

```python
# 열의 기준이 맞지 않아 합칠수 없음 
np.concatenate((x,y),axis=1)
```
